```python
Numbers={"zero": 0, 
         "one" : 1, 
         "two": 2,
        "three": 3,
        "four": 4,
        "five": 5,
        "six": 6,
        "seven": 7,
        "eight": 8,
        "nine": 9,
        "ten": 10,
        "twenty": 20,
        "thirty": 30,
        "forty": 40,
        "fifty": 50,
        "sixty": 60,
        "seventy": 70,
        "eighty": 80,
        "ninety": 90,
        "hundred": 100
        }
Digit= { "single":1,
        "double":2,
        "triple":3,
        "quadruple":4,
        "quintuple":5,
        "sextuple":6,
        "septuple":7,
        "octuple":8,
        "nonuple":9,
        "decuple":10
                      }
Short= {"C M": "CM","P M": "PM","D M": "DM","A M": "AM"}
Money={'DOLLAR':'$','RUPEES':'Rs','DOLLARS':'$'}
    
strng=input()
sent=[]
word=[]
count=0

for i in strng.split(' '):
    sent.append(i)
for j in range(len(sent)):    
    if len(sent[j])>1:
        if Numbers.get(sent[j])!=None : 
            word.append(str(Numbers.get(sent[j])))
        elif Money.get(sent[j])!=None :
             word.append(Money.get(sent[j]))
        elif Digit.get(sent[j])!=None :
            for i in range(Digit.get(sent[j])):
                word.append(sent[j+1])
        else:
            word.append(sent[j])
    #if  len(sent[j])==1:       
    elif len(sent[j])==1 and j+1 < len(sent):
        if len(sent[j+1])==1:
            x=[]
            x.insert(0,sent[j])
            x.insert(1,sent[j+1])
            w=" ".join(x)
            if Short.get(w)!=None:
                word.append(Short.get(w))
            else:
                word.append(sent[j])

    elif len(sent[j-1])==1 and j+1 < len(sent):
        if len(sent[j+1])!=1:
            word.append(sent[j])
    #elif len(sent[j-1])!=1 and len(sent[j+1])!=1:
        #word.append(sent[j])
v=0
for k in range(len(word)):
    if word[k]=='$' or word[k]=='Rs':
        v=word[k-1]
        word[k-1]=word[k]
        word[k]=v

Final=' '.join(word)
    
print(Final)
```
